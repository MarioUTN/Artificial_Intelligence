```python
import random
from datetime import datetime

inicio = datetime(2017, 1, 30)
final =  datetime(2017, 5, 28)

random_date = inicio + (final - inicio) * random.random()

pr['Fecha boardapprovaldate']= pd.date_range(random_date, periods=len(pr), freq='D')
pr
```

```python
pr['Fecha closingdates']= pd.date_range(random_date, periods=len(pr), freq='D')
pr
```

```python
pr['approvalyear'] = pr['Fecha boardapprovaldate'].dt.year
pr['approvalday'] = pr['Fecha boardapprovaldate'].dt.day
pr['approvalweekday'] = pr['Fecha boardapprovaldate'].dt.weekday
pr['closingyear'] = pr['Fecha closingdates'].dt.year
pr['closingday'] = pr['Fecha closingdates'].dt.day
pr['closingweekday'] = pr['Fecha closingdates'].dt.weekday
pr
```
