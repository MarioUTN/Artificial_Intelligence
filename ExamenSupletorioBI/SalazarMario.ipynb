```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Cargar Datos - Dataframe</h2>

```python
dataset = pd.read_csv("test_ETL.csv", encoding='utf-8', sep=',', header=0)
dataset = dataset.head(25).drop_duplicates()
dataset
```

<h2>Histogramas</h2>
